**Question No1:** What is a random variable in probability theory?


**Answer:**

In probability theory, a random variable is a numerical outcome of a random phenomenon or experiment. It maps outcomes of a random process to real numbers, and its value depends on the result of the experiment.

There are two main types of random variables:

Discrete Random Variable:

Takes on a countable number of distinct values (e.g., the number of heads in a series of coin flips, or the number of cars passing through a toll booth in an hour).


Continuous Random Variable:

Takes on an infinite number of possible values, usually within a given range or interval (e.g., the height of people, or the time it takes for a car to travel a certain distance).

A random variable is typically denoted with a capital letter (like X, Y, or Z), and its values can be described by a probability distribution that tells you the likelihood of each possible outcome. For discrete variables, this is often shown with a probability mass function (PMF), and for continuous variables, it's represented with a probability density function (PDF).

**Question No2:** What are the types of random variables?

**Answer:**

There are two main types of random variables in probability theory:

1. Discrete Random Variable

A discrete random variable takes on a countable number of possible values. These values can be finite or countably infinite. The key characteristic of a discrete random variable is that it can be listed or enumerated.

Examples:

Number of heads in a series of coin flips (e.g., 0, 1, 2, ... heads).
Number of cars passing through a toll booth in a given hour (e.g., 0, 1, 2, 3,... cars).
Number of students in a classroom (e.g., 25, 26, 27...).
In the case of discrete random variables, you can often define a probability mass function (PMF), which gives the probability that the random variable takes each of its possible values.

2. Continuous Random Variable

A continuous random variable takes on an uncountably infinite number of values within a certain interval or range. These variables can take any value within a continuous range, and their values are often measured rather than counted.

Examples:

Height of people (e.g., 170.5 cm, 170.51 cm, 170.50001 cm, etc. could all be possible heights).
Time it takes for a car to travel from point A to point B (e.g., 1.5 hours, 1.52 hours, 1.525 hours, etc.).
Temperature at a given location (e.g., 23.1°C, 23.12°C, 23.123°C, and so on).

For continuous random variables, the probability density function (PDF) is used to describe the distribution. The probability that a continuous random variable takes a specific value is always 0; instead, probabilities are associated with intervals or ranges of values.

**Question No3:** What is the difference between discrete and continuous distributions?

**Answer:**

1. Discrete Distribution:

A discrete distribution describes the probability of outcomes of a discrete random variable—one that can take a countable number of values.

Key Features:

Countable outcomes: The random variable can take specific, distinct values (like 0, 1, 2, 3,...).
Probability mass function (PMF): Probabilities are assigned to each individual outcome. The PMF gives the probability that the random variable takes a particular value.
Sum of probabilities equals 1: The total probability of all possible outcomes adds up to 1.

2. Continuous Distribution:

A continuous distribution describes the probability of outcomes of a continuous random variable—one that can take an infinite number of values within a given range or interval.

Key Features:

Uncountable outcomes: The random variable can take any value in a continuous range (e.g., any real number between 0 and 1).
Probability density function (PDF): Instead of assigning a probability to each specific value, a PDF gives the probability density across a range. For any exact value, the probability is 0; instead, the probability is computed over an interval.
Area under the curve: Probabilities are determined by the area under the PDF curve over a range of values. The total area under the curve (for all possible outcomes) equals 1.

**Question No4:** What are probability distribution functions (PDF)?

**Answer:**

A Probability Distribution Function (PDF) describes the likelihood or probability of a random variable taking a particular value (for continuous random variables) or a set of values (for discrete random variables). The concept of a PDF is central to understanding the distribution of probabilities across different possible outcomes for a random variable.

There are two primary types of Probability Distribution Functions:

1. Probability Mass Function (PMF) (for Discrete Random Variables):
For discrete random variables, the PMF gives the probability that the random variable takes a specific value. It assigns probabilities to distinct outcomes.

2. Probability Density Function (PDF) (for Continuous Random Variables):
For continuous random variables, the PDF describes the likelihood of the variable taking a value in a specific range. Unlike a PMF, a PDF does not give the probability of an exact value; rather, it gives the probability density over an interval.

In summary:

* A PMF (for discrete variables) gives the probability for each specific outcome.
* A PDF (for continuous variables) gives the probability density and allows you to compute the probability for a range of outcomes by finding the area under the curve.

**Question No18:** What is the significance of confidence intervals in statistical analysis?

**Answer:**
Confidence intervals are a fundamental concept in statistical analysis, providing a range of values that likely contain the true population parameter (such as the mean, proportion, or difference between means) based on sample data. The significance of confidence intervals lies in the additional information they offer compared to point estimates, as they quantify the uncertainty associated with sample estimates and give a range of plausible values for the parameter of interest.

Key Significance of Confidence Intervals in Statistical Analysis:
Quantifying Uncertainty:

Confidence intervals convey how uncertain we are about the population parameter. For instance, a 95% confidence interval for a population mean suggests that if we were to take many samples and calculate confidence intervals for each, we would expect 95% of them to contain the true population mean.
Without a confidence interval, you only have a single point estimate (e.g., a sample mean), which doesn't provide any information on how much that estimate might vary or how reliable it is.
Providing a Range of Plausible Values:

Confidence intervals give a range of plausible values for a population parameter, which is more informative than a single point estimate.
For example, if the sample mean test score is 80 with a 95% confidence interval of (78, 82), we can say that we are 95% confident that the true population mean lies between 78 and 82. This range accounts for the variability in sample data and provides more useful information for decision-making.
Guiding Decision-Making:

Confidence intervals help in decision-making because they provide an understanding of the possible range of values. For example, in policy-making or business, knowing that a parameter lies within a specific range (with a certain level of confidence) allows more informed decisions.
If a confidence interval for the difference between two treatments does not include zero, this may suggest a significant difference between the two treatments. Conversely, if the interval includes zero, this may indicate no significant difference.
Assessing Statistical Significance:

A confidence interval is closely related to hypothesis testing. If a confidence interval for a parameter (e.g., a population mean) does not contain a value specified in the null hypothesis (e.g., a hypothesized mean of 0), it provides evidence against the null hypothesis.
For example, if the 95% confidence interval for the difference in means between two groups is (1.5, 4.5), we can conclude that there is a significant difference between the groups, because zero is not in the interval.
Interpreting Precision and Reliability:

The width of the confidence interval reflects the precision of the estimate. A narrow interval suggests a more precise estimate of the population parameter, while a wide interval suggests greater uncertainty.
The confidence level (e.g., 90%, 95%, or 99%) indicates the degree of certainty associated with the interval. A higher confidence level leads to a wider interval, while a lower confidence level gives a narrower interval.
Generalizing Results:

Confidence intervals allow researchers to generalize their findings from the sample to the population. They reflect how well the sample estimate approximates the true population parameter and how much variability can be expected.
This is especially important in survey research, clinical trials, and other studies where the goal is to make inferences about a larger population based on sample data.
Examples of Confidence Intervals in Practice:
1. Estimating Population Mean:
Suppose a company wants to estimate the average amount of time it takes to process customer orders. They take a sample of 50 orders and calculate the sample mean processing time to be 30 minutes with a 95% confidence interval of (28, 32) minutes.

This means the company is 95% confident that the true average processing time for all orders in the population is between 28 and 32 minutes.
2. Comparing Two Groups:
In a clinical trial, you compare the effectiveness of two treatments. After analyzing the data, you find that the difference in treatment effectiveness between the two groups is 5%, with a 95% confidence interval of (2%, 8%).

This suggests that the true difference in effectiveness between the two treatments is likely to be between 2% and 8%, and you can be 95% confident that the difference is not zero.

**Question No21:** What are the assumptions for applying the Central Limit Theorem?

**Answer:**
The Central Limit Theorem (CLT) is a fundamental concept in statistics, stating that, for a sufficiently large sample size, the sampling distribution of the sample mean (or sum) will be approximately normal, regardless of the shape of the original population distribution. This allows statisticians to make inferences about population parameters using the normal distribution, even if the original population distribution is not normal.

However, for the CLT to hold accurately, there are certain assumptions that must be met:

Assumptions for Applying the Central Limit Theorem:
Random Sampling:

The samples must be drawn randomly from the population. This ensures that each sample is independent and that the sample is representative of the population.
If the sampling process is biased or non-random, the results may not reflect the true population parameters, violating the conditions of the CLT.
Independence:

The observations within the sample must be independent of each other. This means that the value of one observation should not influence the value of another.
Independence is particularly important when sampling without replacement, as the probability of selecting an item changes with each sample. A common rule of thumb to ensure independence when sampling without replacement is the 10% Rule: The sample size should be no more than 10% of the total population.
Sample Size (Large Enough):

The sample size should be large enough for the CLT to apply. Although the exact size depends on the shape of the population distribution, a common guideline is that a sample size of 30 or more is generally considered large enough for the CLT to hold.
For populations with a severely skewed or non-normal distribution, a larger sample size may be necessary for the sampling distribution to approximate normality. In cases of extreme skewness or outliers, samples of 50 or more may be recommended.
Finite Variance:

The population from which the sample is drawn should have a finite variance. This means that the population data must not have extreme outliers or infinite variance.
If the variance is infinite or undefined (as in some theoretical distributions like Cauchy), the CLT does not apply.
Shape of the Population Distribution (for very small sample sizes):

For very small sample sizes, the shape of the population distribution can affect whether the sampling distribution of the mean is approximately normal. If the population distribution is already normal, the sampling distribution will be normal even for small sample sizes.
If the population distribution is skewed or has heavy tails, the sample size needs to be large enough to invoke the CLT, because the approximation to normality improves as the sample size increases.

Random Sampling: The samples should be selected randomly.
Independence: Sample observations must be independent of one another.

**Question No2:** Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
low, high = 1, 6  # Defining the range of the discrete uniform distribution

# Generate random data (e.g., 1000 samples)
data = np.random.randint(low, high + 1, size=1000)

# Calculate the probability mass function (PMF)
values, counts = np.unique(data, return_counts=True)
pmf = counts / len(data)

# Plot the PMF
plt.bar(values, pmf, width=0.8, align='center', color='skyblue', edgecolor='black')
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(values)
plt.grid(True)
plt.show()
```

**Question No3:** Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

**Answer:**

```python
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome of the Bernoulli trial (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability of the outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1 (inclusive).")

    if x == 1:
        return p
    else:
        return 1 - p

# Example usage:
p = 0.7  # Probability of success
x_values = [0, 1]

for x in x_values:
    print(f"PDF(X = {x}) = {bernoulli_pdf(x, p)}")
```

**Question No6:** Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
a, b = 1, 6  # Defining the range of the uniform distribution

# Values for the CDF calculation
x_values = np.arange(a, b + 1)  # Possible outcomes
cdf_values = (x_values - a + 1) / (b - a + 1)  # CDF formula

# Plot the CDF
plt.step(x_values, cdf_values, where='post', color='skyblue', linewidth=2, label='CDF')

# Title and labels
plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(x_values)
plt.yticks(np.linspace(0, 1, num=11))

# Show the plot
plt.grid(True)
plt.legend()
plt.show()
```

**Question No9:** Write a Python function to calculate Z-scores from a dataset and plot them.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    """
    Calculate the Z-scores for a given dataset.

    Parameters:
    data (array-like): The input dataset (list or numpy array).

    Returns:
    np.array: Z-scores for the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Generate a sample dataset (e.g., from a normal distribution)
data = np.random.normal(0, 1, 1000)  # Mean = 0, Std Dev = 1, 1000 samples

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(10, 6))
plt.plot(z_scores, alpha=0.75, color='skyblue', label='Z-scores')
plt.axhline(0, color='black', linewidth=1, linestyle='--', label='Mean (Z=0)')
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.legend()
plt.grid(True)
plt.show()
```

**Question No10:** Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 100000  # Size of the population
sample_size = 50  # Size of each sample
num_samples = 1000  # Number of samples to draw

# Generate a non-normal population (uniform distribution)
population = np.random.uniform(0, 10, population_size)

# Function to draw samples and compute sample means
def central_limit_theorem(population, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, sample_size)
        sample_means.append(np.mean(sample))
    return sample_means

# Generate sample means
sample_means = central_limit_theorem(population, sample_size, num_samples)

# Plot the histogram of the sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.75, color='skyblue', edgecolor='black')

# Title and labels
plt.title('Central Limit Theorem: Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

# Show the plot
plt.grid(True)
plt.show()
```

**Question No16:**  Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the standard normal distribution PDF
def standard_normal_pdf(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Generate x values (from -4 to 4, typically enough to capture most of the curve)
x_values = np.linspace(-4, 4, 1000)

# Calculate the corresponding y values for the standard normal distribution
y_values = standard_normal_pdf(x_values)

# Plot the standard normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Standard Normal Distribution', color='skyblue', lw=2)

# Title and labels
plt.title('Standard Normal Distribution (Mean = 0, Std Dev = 1)')
plt.xlabel('x')
plt.ylabel('Probability Density')

# Show the plot
plt.grid(True)
plt.legend()
plt.show()
```

**Question No18:** Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to calculate Z-score
def calculate_z_score(data_point, mean, std_dev):
    return (data_point - mean) / std_dev

# Given data
data_point = 2.5  # Example data point
mean = 0  # Mean of the standard normal distribution
std_dev = 1  # Standard deviation of the standard normal distribution

# Calculate the Z-score
z_score = calculate_z_score(data_point, mean, std_dev)

# Calculate the cumulative probability (CDF) for the Z-score
cdf_value = norm.cdf(z_score)

# Print results
print(f"Data Point: {data_point}")
print(f"Z-score: {z_score}")
print(f"Probability (CDF) for Z-score {z_score}: {cdf_value}")

# Plot the standard normal distribution and indicate the Z-score
x = np.linspace(-4, 4, 1000)  # Values from -4 to 4 for the plot
y = norm.pdf(x)  # Standard normal distribution PDF

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='skyblue', lw=2)
plt.fill_between(x, y, where=(x <= z_score), color='skyblue', alpha=0.5, label=f'Z-score: {z_score:.2f}')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
plt.title('Comparison of Z-score with Standard Normal Distribution')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

**Question No19:** Implement hypothesis testing using Z-statistics for a sample dataset.

**Answer:**

```python
import numpy as np
import scipy.stats as stats

# Sample data (e.g., exam scores)
sample_data = np.array([80, 85, 88, 92, 84, 90, 86, 87, 91, 83])

# Population parameters
population_mean = 85  # Known population mean
population_std = 5    # Known population standard deviation

# Sample size
n = len(sample_data)

# Calculate the sample mean
sample_mean = np.mean(sample_data)

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Significance level (e.g., 0.05 for a 95% confidence level)
alpha = 0.05

# Calculate the critical Z-value for a two-tailed test (95% confidence level)
z_critical = stats.norm.ppf(1 - alpha / 2)

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {z_statistic}")
print(f"Critical Z-value: ±{z_critical}")

# Decision: Reject or fail to reject the null hypothesis
if abs(z_statistic) > z_critical:
    print("Reject the null hypothesis (H0). The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0). There is no significant difference.")
```

**Question No20:** Create a confidence interval for a dataset using Python and interpret the result.

**Answer:**

```python
import numpy as np
import scipy.stats as stats

# Sample dataset (e.g., exam scores or some measurements)
sample_data = np.array([80, 85, 88, 92, 84, 90, 86, 87, 91, 83])

# Sample size
n = len(sample_data)

# Sample mean
sample_mean = np.mean(sample_data)

# Sample standard deviation (using ddof=1 for sample standard deviation)
sample_std = np.std(sample_data, ddof=1)

# Confidence level (95%)
confidence_level = 0.95

# Degrees of freedom
df = n - 1

# t-critical value for 95% confidence
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

# Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"T-critical value for {confidence_level*100}% confidence: {t_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"95% Confidence Interval: ({ci_lower}, {ci_upper})")
```

**Question No21:** Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

**Answer:**

```python
import numpy as np
import scipy.stats as stats

# Step 1: Generate random data from a normal distribution
np.random.seed(42)  # Set seed for reproducibility
sample_size = 100  # Number of data points
population_mean = 50  # Assume the population mean
population_std = 10  # Assume the population standard deviation

# Generate the sample data (normally distributed)
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Calculate the sample mean and sample standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for Bessel's correction)

# Step 3: Calculate the 95% confidence interval for the mean
confidence_level = 0.95
alpha = 1 - confidence_level
n = len(sample_data)
df = n - 1  # Degrees of freedom

# t-critical value for the confidence level
t_critical = stats.t.ppf(1 - alpha/2, df)

# Standard error of the mean (SE)
standard_error = sample_std / np.sqrt(n)

# Margin of error
margin_of_error = t_critical * standard_error

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"T-critical value for {confidence_level*100}% confidence: {t_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"95% Confidence Interval: ({ci_lower}, {ci_upper})")
```

**Question No23:** Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

**Answer:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define the parameters of the Poisson distribution
lambda_poisson = 5  # Average rate of events (mean of Poisson distribution)

# Step 2: Generate a range of k values (the number of events)
k_values = np.arange(0, 15)  # Values for k, i.e., number of events

# Step 3: Calculate the CDF for each k-value
cdf_values = poisson.cdf(k_values, mu=lambda_poisson)

# Step 4: Plot the CDF
plt.figure(figsize=(8, 6))
plt.plot(k_values, cdf_values, marker='o', linestyle='-', color='b', label=f'Poisson CDF ($\\lambda$={lambda_poisson})')
plt.fill_between(k_values, cdf_values, color='skyblue', alpha=0.4)

# Add labels and title
plt.title('Cumulative Distribution Function (CDF) of a Poisson Distribution', fontsize=14)
plt.xlabel('k (Number of events)', fontsize=12)
plt.ylabel('CDF', fontsize=12)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

# Step 5: Output specific CDF values for interpretation
for k in k_values:
    print(f"P(X <= {k}) = {poisson.cdf(k, mu=lambda_poisson)}")
```

**Question No27:**  Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a dataset (for illustration purposes)
np.random.seed(42)  # For reproducibility
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation of the distribution
size = 1000  # Number of data points

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Step 2: Calculate the Z-scores and apply Z-score normalization
mean_data = np.mean(data)
std_dev_data = np.std(data)
normalized_data = (data - mean_data) / std_dev_data

# Step 3: Visualize the original and normalized data using histograms
plt.figure(figsize=(12, 6))

# Plot original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Histogram of Normalized Data (Z-score)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()
plt.show()

# Step 4: Print statistics to verify normalization
print(f"Original Data - Mean: {mean_data}, Std Dev: {std_dev_data}")
print(f"Normalized Data - Mean: {np.mean(normalized_data)}, Std Dev: {np.std(normalized_data)}")
```

**Question No28:** Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

**Answer:**

```python
import numpy as np
from scipy.stats import skew, kurtosis

# Function to calculate skewness and kurtosis of a dataset
def calculate_skewness_kurtosis(data):
    # Skewness calculation
    skewness = skew(data)

    # Kurtosis calculation (excess kurtosis, subtracting 3 from the result)
    kurt = kurtosis(data)  # By default, this gives excess kurtosis

    return skewness, kurt

# Step 1: Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation of the distribution
size = 1000  # Number of data points

# Generate the dataset
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Step 2: Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)

# Output the results
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```
